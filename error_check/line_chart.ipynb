```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_students_per_year(csv_file):
    df = pd.read_csv(csv_file)
    summary = df.groupby('Educational Year')['StudentID'].nunique().reset_index(name='Number of Students')
    fig, ax = plt.subplots()
    ax.plot(summary['Educational Year'], summary['Number of Students'], marker='o')
    ax.set_xlabel('Educational Year')
    ax.set_ylabel('Number of Students')
    ax.set_title('Number of Students per Educational Year')
    return fig

plot_students_per_year
```

```python
fig = plot_students_per_year('data/updated_coderminds.csv')
plt.show(fig) 
```
